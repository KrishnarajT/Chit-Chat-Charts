```python
import json
from datetime import datetime
import pandas as pd
import numpy as np
```

```python

with open('Telegram Chat with Jeet.json', 'r') as f:
    data = json.load(f)
```

```python
data['messages']
```

```python
wordDf = pd.DataFrame(columns = ['Speaker', 'Date', 'Word'])
    messageDf = pd.DataFrame(columns = ['Speaker', 'Date', 'Message'])
```

```python
for _, i in enumerate(data['messages']):
    if i['type'] == 'message' and type(i['text']) == str:
        speaker = i['from']
        date = datetime.strptime(i['date'], '%Y-%m-%dT%H:%M:%S')
        message = i['text']
        for j in message.split(' '):
            wordDf = wordDf.append({'Speaker' : speaker, 'Date' : date, 'Word' : j}, ignore_index=True)
        messageDf.loc[_] = [speaker, date, message]
```

```python
messageDf
```

```python
wordDf['Word'].replace('', np.nan, inplace=True)
```

```python
wordDf.dropna(subset=['Word'], inplace=True)
```

```python
wordDf
```

```python
messageDf['Message'].replace('', np.nan, inplace=True)
messageDf.dropna(subset=['Word'], inplace=True)
```

```python
people = set()
for i in wordDf['Speaker']:
    people.add(i)
```

```python

person1, person2 = people
```

```python
wordDf['Speaker'].replace(person1, 1, inplace=True)
wordDf['Speaker'].replace(person2, 2, inplace=True)
```
